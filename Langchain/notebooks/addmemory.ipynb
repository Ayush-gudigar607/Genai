```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
prompts=[
    {
        "role":"user","content":"Hello My name is John. Can you explain the theory of relativity?"},
       { "role":"ai","content":"The theory of relativity explains gravity as space-time curvature."},
       { "role":"user","content":"what is my name?" }
]

res=llm.invoke(prompts)
print(res.content)
```

Manage history of llm

```python
history=[]
while True:
    user_input=input("YOU:")
    if user_input.lower() in ["exit","quit"]:
        break   
    history.append({"role":"user","content":user_input})
    print("User:",user_input)

    res=llm.invoke(history)
    history.append({"role":"ai","content":res.content})
    print("AI:",res.content,"\n")
```
