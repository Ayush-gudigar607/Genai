```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate  # Added the 'g' here
from langchain_core.output_parsers import StrOutputParser
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
out=StrOutputParser()
```

```python
### My Self Chain Node

def transorm_case(result:str):
    return result.upper()
```

```python
prompts=ChatPromptTemplate.from_messages([
    {"role":"system","content":"you are a best translator  and translate input  in {language} language"},
    {"role":"user","content":"{query}"}
])


```

```python
chains= prompts | llm | out | transorm_case
```

```python
res = chains.invoke({
    "language": "Tulu",
    "query": "I am a Python Developer"
})

print(res)





```
