Solution is:Structured output

```python
from pydantic import BaseModel, Field
from typing import List

class ResponseStructure(BaseModel):
    name:str=Field(description="The name of the person extracted from the text")
    location:str=Field(description="The location of the person extracted from the text")
    organization:str=Field(description="The organization of the person extracted from the text")
```

```python
structured_llm=llm.with_structured_output(ResponseStructure)
res=structured_llm.invoke(f"Please extract the names, locations, and organizations from the following text:{text}")
```

```python
res.name, res.location, res.organization
```

```python
res.model_dump()
```

```python
class Movies(BaseModel):
    title:str=Field(description="The title of the movie")
    year:int=Field(description="The year the movie was released")

class AllMovies(BaseModel):
    movies:List[Movies]=Field(description="A list of movies with their titles and release years")

movie_llm=llm.with_structured_output(AllMovies)
```

```python
res=movie_llm.invoke("Provide me 10  recent  movie  in kannada title and its release year.")
for movie in res.movies:
    print(f"{movie.title} - {movie.year}")
```
